# T20 CRICKET WORLD CUP DATA PRE PROCESSING

```python
import pandas as pd
import json
```

1. Process Match Results

```python
with open("t20_json_file/t20_wc_match_results.json") as f:
    data= json.load(f)
```

```python
df_match= pd.DataFrame(data[0]["matchSummary"])
df_match.head()
```

```python
df_match.shape
```

```python
df_match.rename({"scorecard":"match_id"}, axis=1, inplace=True)
```

```python
df_match.head()
```

```python
#Create a match ids dictionary that maps team names to a unique match id. This will be useful later on to link with other tables
Match_id_dict={}
for index, row in df_match.iterrows():
    key1=row["team1"] + " Vs " + row["team2"]
    key2=row["team2"] + " Vs " + row["team1"]
    Match_id_dict[key1]=row["match_id"]
    Match_id_dict[key2]=row["match_id"]
    
Match_id_dict
```

```python
df_match.to_csv('t20_csv_file/dim_match_summary.csv', index= False)
```

2. Process Batting Summary

```python
with open ("t20_json_file/t20_wc_batting_summary.json") as f:
    data= json.load(f)
    
    all_record=[]
    
    for rec in data:
        all_record.extend(rec["battingSummary"])
df_batting= pd.DataFrame(all_record)
```

```python
df_batting.head(11)
```

```python
df_batting["Out/Not_out"]=df_batting.dismissal.apply(lambda x:"Out" if len(x)>0 else "Not_out")
```

```python
df_batting.drop(columns=["dismissal"], inplace=True)
df_batting.head()
```

```python
df_batting["batsmanName"].unique()
```

```python
#Cleaning up weird Charters for the name
df_batting["batsmanName"]=df_batting['batsmanName'].apply(lambda x:x.replace("â€",""))
df_batting["batsmanName"]=df_batting['batsmanName'].apply(lambda x:x.replace("xa0",""))
```

```python
df_batting.head(10)
```

```python
df_batting["match_id"]= df_batting["match"].map(Match_id_dict)

df_batting.head()
```

```python
df_batting.shape
```

```python
df_batting.to_csv("t20_csv_file/fact_bating_summary.csv", index=False)
```

3. Processing Bowling Summary

```python
with open("t20_json_file/t20_wc_bowling_summary.json") as f:
    data=json.load(f)
    all_record=[]
    for rec in data:
        all_record.extend(rec["bowlingSummary"])
all_record
```

```python
df_bowling=pd.DataFrame(all_record)
print(df_bowling.shape)
df_bowling.head()
```

```python
df_bowling["Match_id"]=df_bowling["match"].map(Match_id_dict)
df_bowling.head()
```

```python
df_bowling.to_csv("t20_csv_file/fact_bowling_summary.csv", index=False)
```

4. Process Player Information

```python
with open("t20_json_file/t20_wc_player_info.json") as f:
    data= json.load(f)
    
df_players= pd.DataFrame(data)
print(df_players.shape)

df_players.head(10)
```

```python
df_players["name"]
```

```python
#Cleaning up the weird charters
df_players["name"]= df_players['name'].apply(lambda x: x.replace("â€"," "))
df_players["name"]= df_players['name'].apply(lambda x: x.replace("†"," "))
df_players["name"]= df_players['name'].apply(lambda x: x.replace("\xa0"," "))
```

```python
df_players.head(10)
```

```python
df_players[df_players["team"]=="India"]
```

```python
df_players.to_csv("t20_csv_file/dim_players.csv", index= False)
```
